```python
import pandas as pd
import gdown
import os
```

```python
df = pd.read_excel("test_set.xlsx")

df["uid"] = df["Property Name"] + "_" + df["Property Address"]

read = 0
success = 0
failed = 0

for index, row in df.iterrows():
    google_drive_url = row["Link to OM"]
    filename = row["uid"]

    output_folder = r'C:\Users\Jackson Wexler\Documents\Repository\RobbyGPT\OM_Repo'
    output_filename = filename + ".pdf"

    # Construct the full output path
    output_path = os.path.join(output_folder, output_filename)

    try:
        # Download the file
        gdown.download(google_drive_url, output_path, quiet=False, fuzzy=True)

    except Exception as e:
        print(f"Error downloading file: {e}")
        failed += 1
    
    else:
        print(f"File downloaded to: {output_path}")
        success += 1

    finally:
        read += 1
        print(f"Read: {read}, Success: {success}, Failed: {failed}")
```
